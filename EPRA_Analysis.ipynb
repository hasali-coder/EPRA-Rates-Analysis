```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("EPRA_SEPT.xlsx")
df.head()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Years', y='Fuel costs EPRA KES/LTS')
plt.title('Fuel Costs EPRA KES/LTS over the Years')
plt.xlabel('Year')
plt.ylabel('Fuel Costs (KES/LTS)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Years', y='Employees Costs%')
plt.title('Employees Costs Percentage over the Years')
plt.xlabel('Year')
plt.ylabel('Employees Costs (%)')
plt.show()

```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Years', y='Intrest Borrowing %')
plt.title('Interest Borrowing Percentage over the Years')
plt.xlabel('Year')
plt.ylabel('Interest Borrowing (%)')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Select the columns for the multiple lines chart
columns = ['EPRA RATES (KES/LTRS)- WITHIN TOWN', 'Fuel costs EPRA KES/LTS', 'Operational Costs KES/KM']

# Create a figure and axis
plt.figure(figsize=(12, 6))
ax = plt.subplot(111)

# Plot each line with a label
for col in columns:
    plt.plot(df['Years'], df[col], marker='o', label=col)

# Add labels, title, legend, and grid
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Multiple Lines Chart with Legends and Data Points')
plt.legend(loc='upper left')
plt.grid(True)

# Show data points on the lines
for col in columns:
    for x, y in zip(df['Years'], df[col]):
        plt.text(x, y, f'{y:.2f}', ha='right', va='bottom')

# Show the chart
plt.show()
```

```python

# Select the columns for the comparison
columns = ['Intrest Borrowing %', 'Operationa Costs incl diesel KES/KMs']

# Create a figure and axes
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot a bar chart for "Intrest Borrowing %"
bar_width = 0.4
index = np.arange(len(df['Years']))
bar1 = ax1.bar(index, df['Intrest Borrowing %'], bar_width, label='Interest Borrowing %', align='center', alpha=0.7)

# Set axis labels and title for the bar chart
ax1.set_xlabel('Year')
ax1.set_ylabel('Interest Borrowing %', color='tab:blue')
ax1.set_title('Comparison of Interest Borrowing % and Borrowing Cost')
ax1.set_xticks(index)
ax1.set_xticklabels(df['Years'])
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Create a second y-axis for the line chart
ax2 = ax1.twinx()

# Plot a line chart for "Operational Costs incl diesel KES/KMs"
line2 = ax2.plot(index, df['Operationa Costs incl diesel KES/KMs'], marker='o', color='tab:red', label='Borrowing Cost')
ax2.set_ylabel('Borrowing Cost (KES/KMs)', color='tab:red')
ax2.tick_params(axis='y', labelcolor='tab:red')

# Combine the legend for both charts
lines = [bar1[0], line2[0]]
labels = [line.get_label() for line in lines]
ax1.legend(lines, labels, loc='upper left')

# Add data points for the line chart
for x, y in zip(index, df['Operationa Costs incl diesel KES/KMs']):
    ax2.text(x, y, f'{y:.2f}', ha='right', va='bottom')

# Show the combined chart
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


X = df['Years'].values.reshape(-1, 1)
y = df['EPRA RATES (KES/LTRS)- WITHIN TOWN'].values

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Make predictions for the same years
years_to_predict = df['Years'].values.reshape(-1, 1)
predictions = model.predict(years_to_predict)

# Create a line graph to visualize the predictions
plt.figure(figsize=(10, 6))
plt.plot(df['Years'], df['EPRA RATES (KES/LTRS)- WITHIN TOWN'], marker='o', label='Actual', color='blue')
plt.plot(df['Years'], predictions, marker='x', label='Predicted', color='red')
plt.xlabel('Years')
plt.ylabel('EPRA RATES (KES/LTRS)- WITHIN TOWN')
plt.title('Actual vs. Predicted EPRA RATES (KES/LTRS)- WITHIN TOWN')
plt.legend()
plt.grid(True)

# Show the graph
plt.show()

```

```python
import matplotlib.pyplot as plt

# Multiply the columns by 100 to remove decimal places
df['Intrest Borrowing %'] *= 100
df['Employees Costs%'] *= 100

# Create a figure and axis
plt.figure(figsize=(10, 6))
ax = plt.subplot(111)

# Plot "Interest Borrowing %" and "Employees Costs%" lines with labels
plt.plot(df['Years'], df['Intrest Borrowing %'], marker='o', label='Interest Borrowing %')
plt.plot(df['Years'], df['Employees Costs%'], marker='o', label='Employees Costs %')

# Add labels, title, legend, and grid
plt.xlabel('Year')
plt.ylabel('%')
plt.title('Interest Borrowing % and Employees Costs % Over the Years')
plt.legend()
plt.grid(True)

# Show data points on the lines
for col in ['Intrest Borrowing %', 'Employees Costs%']:
    for x, y in zip(df['Years'], df[col]):
        plt.text(x, y, f'{y:.0f}', ha='right', va='bottom')

# Show the chart
plt.show()
```
